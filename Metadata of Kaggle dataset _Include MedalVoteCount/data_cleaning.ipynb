```python
import pandas as pd
import numpy as np


df= pd.read_csv('medals.csv')
df.head()
```

```python
print(df.isna().sum())
df.info()
```

```python
df.nunique()
```

```python
df['Date'] = pd.to_datetime(df['ct'])
df['Date'] = df['Date'].dt.to_period('M')
df.set_index(df['Date'],inplace=True)
```

```python
df['URL'] = df['dataUrl'].apply(lambda x: x.split('/')[0])

URL_count = df.groupby('URL')['URL'].agg('count').sort_values(ascending=False)
URL_count
```

```python
len(URL_count[URL_count<5])
```

```python
len(URL_count[URL_count>=5])
```

```python
URL_less_than_5 = URL_count[URL_count<5]
URL_less_than_5
```

```python
df["Website"] = df['URL'].apply(lambda x: 'Other' if x in URL_less_than_5 else x)
```

```python
df.head()
```

```python
drop_columns =['dataUrl','ct', 'title' ,'Date', 'description','key' ,'license','URL','totalkernel']
data = df.drop(drop_columns,axis=1)
```

```python
data.rename(columns={'DataSetMedals':'Medal',
                      'totalviews':'Views',
                        'votecount':'Total_vote',
                        'medalvotecount':'Medal_vote',
                        'totaldownloads':'Downloads'},inplace=True)
```
